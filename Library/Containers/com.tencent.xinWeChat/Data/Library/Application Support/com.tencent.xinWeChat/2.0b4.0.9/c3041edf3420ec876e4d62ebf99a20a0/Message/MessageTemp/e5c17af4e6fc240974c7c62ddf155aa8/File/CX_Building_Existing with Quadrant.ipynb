```python
# import library
import pandas as pd
import numpy as np
import scipy.stats as stats
import warnings
warnings.filterwarnings('ignore')
```

```python
#load Original_Data2021 Dataset
df_cx = pd.read_excel('Original_Data2021.xlsx',engine='openpyxl',index_col=0, sheet_name = 'CX 2021')  
```

```python
df_cx.columns = df_cx.iloc[0]
```

```python
df_cx = df_cx.iloc[1: , :]
```

```python
df_cx
# deal with NaN value
df_cx.fillna(0, inplace=True)
```

```python
# Calculate Sentiment
pos_percent = df_cx['positive']/(df_cx['positive']+df_cx['negative'])
Sentiment = (pos_percent-0.5)/0.5
col_name = df_cx.columns.tolist()
df_cx.insert(col_name.index('negative')+1, "Sentiment", Sentiment) 
df_cx.fillna(0, inplace=True)
```

```python
# Select useful coloumns
cx = pd.DataFrame(df_cx,columns = ['Company Name','Category','FB_Posts','Ins_Posts','Youtube_Posts','Tweets','Twitter conversation rate',
                        'Customized news feed (0/1)','Click to call (0/1)',
                        'web link to app on Android (0/1) ','Easy access to mobile transaction (0/1)',
                         'FB_Likes','FB_Shares',
                                     'FB_Engagement',
                                     'Ins_Likes',
                                     'Tw_Likes',
                                     'Retweets',
                                     'Tw_Engagement',
                                     'tweet',
                                     'Youtube_Likes',
                                     'Sentiment',
                                     'IOS rating',
                                     'iOS positive sentiment ',
                                     'Android rating',
                                     'Android positive sentiment',
                                     'trf_email',
                                     'trf_social'])
```

### Company_initiated Part

```python
# create founction to calculate z-score in company_initiated part
def company_initiated(df):
    # calculate zscore of FB_Posts
    FB_Posts_zscore = stats.zscore(df['FB_Posts'])

    # insert 'FB_Posts_zscore' after 'FB_Posts' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('FB_Posts')+1, 'FB_Posts_zscore', FB_Posts_zscore)
    
    # calculate zscore of Ins_Posts
    Ins_Posts_zscore = stats.zscore(df['Ins_Posts'])

    # insert 'Ins_Posts_zscore' after 'Ins_Posts' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Ins_Posts')+1, 'Ins_Posts_zscore', Ins_Posts_zscore)
    
    # calculate zscore of Tweets
    Tweets_zscore = stats.zscore(df['Tweets'])

    # insert 'Tweets_zscore' after 'Tweets' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Tweets')+1, 'Tweets_zscore', Tweets_zscore)
    
    # calculate zscore of Twitter conversation rate
    Twitter_conversation_rate_zscore = stats.zscore(df['Twitter conversation rate'])

    # insert 'Android_App_Rating_zscore' after 'Performance_Android_App_rating' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Twitter conversation rate')+1, 'Twitter_conversation_rate_zscore', Twitter_conversation_rate_zscore)
    
    # calculate zscore of Youtube_Posts
    Youtube_Posts_zscore = stats.zscore(df['Youtube_Posts'])

    # insert 'Youtube_Posts_zscore' after 'Performance_Android_App_rating' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Youtube_Posts')+1, 'Youtube_Posts_zscore', Youtube_Posts_zscore)
    
    #insert "sum of binary"
    Sum_Binary = df['Customized news feed (0/1)']+df['Click to call (0/1)']+df['web link to app on Android (0/1) ']+df['Easy access to mobile transaction (0/1)']
    col_name = df.columns.tolist()
    df.insert(col_name.index('Easy access to mobile transaction (0/1)')+1, "Sum_Binary", Sum_Binary)
    
    #calulate zscore of Sum_Binary
    Mobile_Capilities_zscore = stats.zscore(df['Sum_Binary'])
    
    #insert 'Mobile_Capilities_zscore' after 'Sum_Binary'
    col_name = df.columns.tolist()
    df.insert(col_name.index('Sum_Binary')+1, "Mobile_Capilities_zscore", Mobile_Capilities_zscore) 
    
    X_axis = 0.4*(df['FB_Posts_zscore']+df['Ins_Posts_zscore']+df['Tweets_zscore']
                  +df['Youtube_Posts_zscore'])+0.6*(df['Twitter_conversation_rate_zscore']+df['Mobile_Capilities_zscore'])
    
    # insert 'X_axis' after 'Android_POS_PERCENT_zscore' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Mobile_Capilities_zscore')+1, 'X_axis', X_axis)
```

### Customer_initiated Part

```python
# create founction to calculate z-score in customer_initiated part
def customer_initiated(df):
    # calculate zscore of FB_Likes
    FB_Likes_zscore = stats.zscore(df['FB_Likes'])

    # insert 'FB_Likes_zscore' after 'FB_Likes' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('FB_Likes')+1, 'FB_Likes_zscore', FB_Likes_zscore)
    
    # calculate zscore of FB_Shares
    FB_Shares_zscore = stats.zscore(df['FB_Shares'])

    # insert 'FB_Shares_zscore' after 'FB_Shares' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('FB_Shares')+1, 'FB_Shares_zscore', FB_Shares_zscore)
    
    # calculate zscore of FB_Engagement
    FB_Engagement_zscore = stats.zscore(df['FB_Engagement'])

    # insert 'FB_Engagement_zscore' after 'FB_Engagement' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('FB_Engagement')+1, 'FB_Engagement_zscore', FB_Engagement_zscore)
    
    # calculate zscore of Ins_Likes
    Ins_Likes_zscore = stats.zscore(df['Ins_Likes'])

    # insert 'Ins_Likes_zscore' after 'Ins_Likes' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Ins_Likes')+1, 'Ins_Likes_zscore', Ins_Likes_zscore)
    
    # calculate zscore of Tw_Likes
    Tw_Likes_zscore = stats.zscore(df['Tw_Likes'])

    # insert 'Tw_Likes_zscore' after 'Tw_Likes' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Tw_Likes')+1, 'Tw_Likes_zscore', Tw_Likes_zscore)
    
    # calculate zscore of Retweets
    Retweets_zscore = stats.zscore(df['Retweets'])

    # insert 'Retweets_zscore' after 'Retweets' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Retweets')+1, 'Retweets_zscore', Retweets_zscore)
    
    # calculate zscore of Tw_Engagement
    Tw_Engagement_zscore = stats.zscore(df['Tw_Engagement'])

    # insert 'Tw_Engagement_zscore' after 'Tw_Engagement' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Tw_Engagement')+1, 'Tw_Engagement_zscore', Tw_Engagement_zscore)
    
    # calculate zscore of tweet
    tweet_zscore = stats.zscore(df['tweet'])

    # insert 'tweet_zscore' after 'tweet' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('tweet')+1, 'tweet_zscore', tweet_zscore)
    
    # calculate zscore of Youtube_Likes
    Youtube_Likes_zscore = stats.zscore(df['Youtube_Likes'])

    # insert 'Youtube_Likes_zscore' after 'Youtube_Likes' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Youtube_Likes')+1, 'Youtube_Likes_zscore', Youtube_Likes_zscore)
    
    # calculate zscore of Sentiment
    Sentiment_zscore = stats.zscore(df['Sentiment'])

    # insert 'Sentiment_zscore' after 'Sentiment' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Sentiment')+1, 'Sentiment_zscore', Sentiment_zscore)
    
    # calculate zscore of IOS rating
    IOS_rating_zscore = stats.zscore(df['IOS rating'])

    # insert 'IOS rating_zscore' after 'IOS rating' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('IOS rating')+1, 'IOS_rating_zscore', IOS_rating_zscore)
    
    # calculate zscore of iOS positive sentiment 
    iOS_positive_sentiment_zscore = stats.zscore(df['iOS positive sentiment '])

    # insert 'iOS positive sentiment _zscore' after 'iOS positive sentiment ' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('iOS positive sentiment ')+1, 'iOS_positive_sentiment_zscore', iOS_positive_sentiment_zscore)
    
    # calculate zscore of Android rating
    Android_rating_zscore = stats.zscore(df['Android rating'])

    # insert 'Android rating_zscore' after 'Android rating' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Android rating')+1, 'Android_rating_zscore', Android_rating_zscore)
    
    # calculate zscore of Android positive sentiment
    Android_positive_sentiment_zscore = stats.zscore(df['Android positive sentiment'])

    # insert 'Android positive sentiment_zscore' after 'Android positive sentiment' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Android positive sentiment')+1, 'Android_positive_sentiment_zscore', Android_positive_sentiment_zscore)
    
    # calculate zscore of trf_email
    trf_email_zscore = stats.zscore(df['trf_email'])

    # insert 'trf_email_zscore' after 'trf_email' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('trf_email')+1, 'trf_email_zscore', trf_email_zscore)
    
    # calculate zscore of trf_social
    trf_social_zscore = stats.zscore(df['trf_social'])

    # insert 'trf_social_zscore' after 'trf_social' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('trf_social')+1, 'trf_social_zscore', trf_social_zscore)
    
    Y_axis = 0.6*(df['FB_Likes_zscore']+df['FB_Shares_zscore']+df['FB_Engagement_zscore']+df['Ins_Likes_zscore']+
                            df['Tw_Likes_zscore']+df['Retweets_zscore']+df['Tw_Engagement_zscore']+
                            df['tweet_zscore']+df['Youtube_Likes_zscore']+df['Sentiment_zscore']+
                            df['IOS_rating_zscore']+df['iOS_positive_sentiment_zscore']+df['Android_rating_zscore']+
                            df['Android_positive_sentiment_zscore'])+0.4*(df['trf_email_zscore']+df['trf_social_zscore'])
    
    
    # insert 'Y_axis' after 'trf_social_zscore' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('trf_social_zscore')+1, 'Y_axis', Y_axis)
```

### Final Score & Final Ranking

```python
# create founction to calculate final score and ranking
def final_score_ranking(df):
    # claculate final score
    Final_Score = df['X_axis'] + df['Y_axis']       
        
    # insert 'Final_Score' after 'Y_axis' column
    col_name = df.columns.tolist()
    df.insert(col_name.index('Y_axis')+1, 'Final_Score', Final_Score)
    
    # rank the final score and label the ranking number
    df["Rank"] = df["Final_Score"].rank(method='first',ascending=False )
```

```python
##Quadrant
def Quadrant(df):
  #set the length
  Overall_Comp_position = [0] * len(df)
  Overall_Comp_recommendation = [0] * len(df)
  for i in range(len(df)):
      if df.iloc[i]['X_axis'] > 0 and df.iloc[i]['Y_axis'] > 0:
          Overall_Comp_position[i] = 'Leader'
          Overall_Comp_recommendation[i] = "Maintain and calibrate the firm's existing cx strategy. Focus on advocacy through sustained engagement with customers."
      elif df.iloc[i]['X_axis'] > 0 and df.iloc[i]['Y_axis'] < 0:
          Overall_Comp_position[i] = 'Hard Worker'
          Overall_Comp_recommendation[i] = "Collaborate with the firm's audience, establish best practices to better understand their needs and calibrate content to drive conversion."
      elif df.iloc[i]['X_axis'] < 0 and df.iloc[i]['Y_axis'] < 0:
          Overall_Comp_position[i] = 'Follower'
          Overall_Comp_recommendation[i] = "Reengineer the firm's customer experience strategy. Perform research to understand jobs-to-be-done, segment the audience and calibrate content to drive conversion."
      else:
          Overall_Comp_position[i] = 'Star'
          Overall_Comp_recommendation[i] = "Strengthen key capabilities that drive a contemporary customer experience strategy - context, content, conversion, calibration and collaboration, at each stage of the customer jouney."
  col_name = df.columns.tolist()
  df.insert(col_name.index('Y_axis')+1, 'Overall_Comp_position', Overall_Comp_position)
  # insert 'Overall_Comp_recommendation' after 'Website_Ret_recommendation' column
  df.insert(col_name.index('Y_axis')+2, 'Overall_Comp_recommendation', Overall_Comp_recommendation)
```
